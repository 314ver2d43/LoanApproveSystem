
The dataset contains 13 features: 8 categorical and 5 numerical.

Most features are string-based, but financial indicators are represented by numbers. The target feature Loan_Status is categorical, as it takes values Y or N.

FEATURE DESCRIPTION:

Loan_ID - application ID

Gender - applicant's gender

Married - marital status

Dependents - number of family members

Education - education level

Self_Employed - employment status

ApplicantIncome - applicant's income

CoapplicantIncome - co-applicant's income

LoanAmount - requested loan amount

Loan_Amount_Term - loan term

Credit_History - credit history availability 1 - yes 0 - no

Property_Area - location

Loan_Status - loan approval (target feature)
